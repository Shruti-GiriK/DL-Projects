# **Text Classification - Understanding the Sentiment by The Reviews**

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

## **Data Understanding**

```python
df=pd.read_csv('/content/cleaned_reviews.csv')
df
```

```python
df.info()
```

## **Data Cleaning**

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
print(df[df.duplicated()])
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df[df.duplicated()])
```

```python
df
```

```python
df.describe()
```

```python
df.drop(['review_score','cleaned_review_length'],axis=1,inplace=True)
```

```python
df['sentiments'].replace(['Negative','Neutral','Positive'],[1,2,3],inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(df['cleaned_review'],df['sentiments'],test_size=0.2,random_state=42)
```

```python
vectorizer=TfidfVectorizer()
xtrain_vec=vectorizer.fit_transform(xtrain)
xtest_vec=vectorizer.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain_vec,ytrain)
```

```python
model.predict(vectorizer.transform(['Its  a bad product']))
```

```python
model.predict(vectorizer.transform(['Its an okay-ish product']))
```

```python
model.predict(vectorizer.transform(['I love this product, Its soo nice !!']))
```

```python
model.score(xtest_vec,ytest)
```
